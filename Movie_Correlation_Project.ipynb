```python
import pandas as pd
import seaborn as sns 
import numpy as np 
import matplotlib
import matplotlib.pyplot as plt 
plt.style.use('ggplot')
from matplotlib.pyplot import figure 

%matplot inline 
matplotlib.rcParams['figure.figsize'] = (12.8)
```

```python
# read in the data 
df = pd.read_csv(r"C:\Users\blood\OneDrive\Documents\Data_Analysis_Projects\movies.csv")
```

```python
# let's look at the data, first 10 rows of the data 
df.head(10)
```

```python
# Check if there is null value in each column and the data type 
df.info()
```

```python
df.dropna()
df.info()
```

```python
# Remove missing data 
df = df.dropna()
```

```python
# Change data type 
df['budget'] = df['budget'].astype('int64')
df['gross'] = df['gross'].astype('int64')
df.head(10)
```

```python
# Create correct Year Column 
df['year'] = df['released'].astype(str).str[:4]
```

```python
# Sort the data by gross in descending order 
df = df.sort_values(by = ['gross'], inplace = False, ascending = False)
```

```python
# Remove Dulpicate 
df = df.drop_duplicates()
df 
```

```python
# Budget High Correlation 
# Company high Correlation 
plt.figure(figsize=(20, 10))
plt.scatter(x=df['budget'], y=df['gross'])
plt.title('Budget vs Gross Earnings')
plt.xlabel('Gross Earning')
plt.ylabel('Budge from Film')
plt.show()
```

```python
# Plot budget vs gross using seaborn 
plt.figure(figsize=(20, 10))
sns.regplot(x='budget',y='gross',data =df, scatter_kws={"color":"red"},line_kws={"color":"blue"})
```

```python
df.corr()#pearson, kendall, spearman 
```

```python
correlation_matrix = df.corr()
plt.figure(figsize=(20, 10))
sns.heatmap(correlation_matrix, annot=True)
plt.title('Corrleation Matrix for Numeric Features')
plt.xlabel('Movie Features')
plt.ylabel('Movie Features')
plt.show()
```

```python
# Look at Company 

df_numerized = df 

for col_names in df_numerized.columns:
    if(df_numerized[col_names].dtype =='object'):
        df_numerized[col_names] = df_numerized[col_names].astype('category')
        df_numerized[col_names] = df_numerized[col_names].cat.codes
df_numerized
```

```python
correlation_matrix = df_numerized.corr()
plt.figure(figsize=(20, 10))
sns.heatmap(correlation_matrix, annot=True)
plt.title('Corrleation Matrix for Numeric Features')
plt.xlabel('Movie Features')
plt.ylabel('Movie Features')
plt.show()
```

```python
corr_pairs = correlation_matrix.unstack()
corr_pairs = corr_pairs.sort_values()
pd.set_option('display.max_rows', 300)
sort_pairs = corr_pairs[(corr_pairs) >0.5]
sort_pairs
```
